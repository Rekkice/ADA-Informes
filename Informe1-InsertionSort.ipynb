# 2. Insertion sort
## 2.1 Descripci√≥n del algoritmo
Este algoritmo recibe una lista de largo n como entrada. Luego, ordena sus elementos al iterar sobre la lista y moverlos a la posici√≥n que les corresponde. Es similar a la forma en que se ordenan las cartas en una mano.

1. Se asume que el primer dato ya est√° ordenado.
2. Cada dato se compara con los que est√°n a su izquieda. Luego, si el dato es menor, se intercambian su posiciones, hasta que est√© en una posici√≥n en donde el dato a su izquierda es menor.
3. Se sigue iterando hasta que la lista est√© completamente ordenada.

## 2.2 C√≥digo

```python
def insertion_sort(lista, verbose=False):
  if verbose: print(f"Entrada: {lista}")
  comparaciones = 0

  for i in range(1, len(lista)): #empieza en 1 porque se asume que el primer elemento ya est√° ordenado
    if verbose: print("Se elige el dato no ordenado n¬∞", i, "\n")

    j = i # la comparaci√≥n comienza en la posici√≥n anterior al elemento
    while ( (j-1 >= 0) and (lista[j] < lista[j - 1]) ): # se compara el elemento con los que est√°n a su izquierda. si es mayor, se intercambian sus posiciones
      if verbose: print(f"El elemento {lista[i]} se intercambia con el elemento {lista[j - 1]}")
      lista[j], lista[j - 1] = lista[j - 1], lista[j]
      j -= 1
      comparaciones += 1

    if verbose: print(lista, "\n")

  return lista, comparaciones

print("Lista ordenada:",insertion_sort([3,5,2,1])[0])
```

##2.3 Ejemplo
Consideremos la *lista* [3, 5, 2, 1] como ejemplo.

En la primera pasada, el dato 5 se queda en su posici√≥n porque ya es menor a 3.

*lista* = [3, 5, 2, 1]

En la segunda pasada, 2 se intercambia con 5 y luego 3 para quedar en la primera posici√≥n de la lista.

*lista* = [2, 3, 5, 1]

Finalmente, en la tercera pasada, 1 se intercambia con 5, 3 y 2 para quedar en la primera posici√≥n de la lista.

*lista* = [1, 2, 3, 5]

## 2.4 Ejecuci√≥n del c√≥digo paso a paso

# 4. Correctitud
## Teorema (Correctitud).
El algoritmo Insertion Sort retorna una lista con los mismos elementos de entrada, pero ordenados de menor a mayor.

## Prueba del teorema
Se puede observar que en cada iteraci√≥n de Insertion Sort, se agrega un elemento ordenado antes del elemento en la posici√≥n *i*. Por lo que probaremos la siguiente propiedad invariante de bucle para el Insertion Sort:

> Al comienzo de cada iteraci√≥n  del primer bucle, los *i - 1* primeros elementos corresponden a elementos ordenados de menor a mayor.

## Inicializaci√≥n

Al comienzo de la primera iteraci√≥n (i = 1), los primeros *i - 1* elementos del arreglo cumplen esta propiedad. Se asume que el primer elemento ya est√° ordenado, por lo tanto, se cumple esta propiedad.

## Mantenci√≥n

Aplicando inducci√≥n, asumimos que al comienzo de cada iteraci√≥n se cumple la propiedad. Los *i - 1* primeros elementos del arreglo corresponden a elementos ordenados de menor a mayor.

Consideremos a X como el elemento que est√° en la posici√≥n *i* al comienzo de la *i*-√©sima iteraci√≥n.

Durante la iteraci√≥n, X intercambia posiciones entre los elementos ya ordenados hasta que se cumpla la condici√≥n de que el valor del siguiente elemento sea menor que el valor de X.

Al finalizar la iteraci√≥n, X queda en una posici√≥n menor a *i* y se seguir√° cumpliendo la propiedad invariante.

## Correctitud
La propiedad de bucle invariante es verdadera al inicio del bucle y se mantiene en cada iteraci√≥n. Por lo tanto, podemos decir que al finalizar la n-√©sima iteraci√≥n del algoritmo, el resultado ser√° una lista con los mismos elementos del arreglo de entrada pero ordenados de menor a mayor.

Ahora realizaremos un numero de experimentos que nos ayudaran a visualizar la eficacia del insertion sort en ciertos aspectos.

##5.1. Numero de comparaciones

En este experimento se comprobara el n√∫mero de comparaciones que el insertion sort realiza poder ordenar una lista en su totailidad.

Para ello se hizo una grafica que muestra:

* La cantidad de comparaciones del peor caso calculadas matem√°ticamente,
* La cantidad de comparaciones del mejor caso calculadas matem√°ticamente (n-1 comparaciones si el arreglo est√° ordenado)
* La cantidad de comparaciones realizadas experimentalmente para entradas aleatorias

Para un problema que tiene un tama√±o $n \in [5,20]$

```python
import matplotlib.pyplot as plt
import random

x=[n for n in range(5,21)] 
y1=[n*(n-1)/2 for n in range(5,21)] # worst case
y2=[n-1 for n in range(5,21)] # best case
y=[]; 

for n in range(5,21):
  a = random.sample(range(1, 100), n)
  a,counter = insertion_sort(a)
  y.append(counter)

plt.plot(x,y)
plt.plot(x,y1)
plt.plot(x,y2)
plt.legend(["Insertion Sort", "theoretical worst case", "theoretical best case"])

plt.xlabel('n')
plt.ylabel('number of operations')
plt.show()
```

### An√°lisis de resultados

Como se puede apreciar en el gr√°fico, el n√∫mero de comparaciones emp√≠rico cumple la expectativa, de realizar menos comparaciones que el peor caso pero mas que el mejor caso.

De hecho, el tiempo emp√≠rico esta bastante cercano a lo que seria la mediana de ambas proyecciones.

##5.2. Tiempo de ejecuci√≥n (experimental)

En el siguiente experimento se va a comprobar si el algoritmo cumple con el tiempo promedio esperado, que ser√≠a un tiempo $O(n^2)$.


```python
import matplotlib.pyplot as plt
import datetime
from timeit import repeat

x=[]; y=[]

for n in range(5,500):

  a = random.sample(range(1, 1000), n)

  #la funci√≥n repeat est√° ejecutando 20 veces insertion_sort con el arreglo a y retorna el mejor tiepo obtenido.
  t = repeat(setup="from __main__ import insertion_sort", stmt=f"insertion_sort({a})", repeat=1, number=10)

  x.append(n)
  y.append(t)


plt.plot(x,y)
plt.legend(["Insertion Sort"])

plt.xlabel('n')
plt.ylabel('time in ms')
plt.show()
```
